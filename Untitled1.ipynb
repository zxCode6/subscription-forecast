```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv('train.csv')
train.head()
```

```python
test = pd.read_csv('test.csv')
test.head()
```

```python
train.drop(columns='id',inplace=True)
test.drop(columns='id',inplace=True)
```

```python
y = train['subscribe']
train.drop(columns='subscribe',inplace=True)
```

```python
X_concat = pd.concat([train, test],axis=0)
```

```python
X_concat.shape
```

```python
train.shape,test.shape
```

```python
X_concat.head()
```

```python
y = y.map({'no':0, 'yes':1})
y.head()
```

```python
num_columns = X_concat.select_dtypes(exclude='O').columns
X_concat[num_columns]
```

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
X_concat[num_columns] = min_max_scaler.fit_transform(X_concat[num_columns])
X_concat[num_columns]
```

```python
X_concat['education'] = X_concat['education'].map({'unknown': 0, 'illiterate': 1, 'basic.4y': 2, 'basic.6y': 3,
                    'basic.9y': 4, 'high.school': 5, 'university.degree': 6, 'professional.course': 7})

X_concat['month'] = X_concat['month'].map({'mar': 3, 'apr': 4, 'may': 5, 'jun': 6, 'jul': 7, 'aug': 8,
                 'sep': 9, 'oct': 10, 'nov': 11, 'dec': 12})
X_concat['day_of_week'] = X_concat['day_of_week'].map({'mon': 0, 'tue': 1, 'wed': 2, 'thu': 3, 'fri': 4})
```

```python
X_concat.info()
```

```python
X_concat = pd.get_dummies(X_concat)
X_concat.head()
```

```python
x_train = X_concat.iloc[train.index]
x_test = X_concat.iloc[test.index]
x_train.shape,x_test.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import GridSearchCV

# 设置随机森林回归模型的参数
param_grid = {
    'n_estimators': [10, 50, 100, 200],
    'max_depth': [10, 20, 30, 50],
}

# 网格搜索法确定最佳参数
rf = RandomForestClassifier(random_state=42)
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5)
grid_search.fit(x_train, y)

# 输出最佳参数
print(grid_search.best_params_)
```

```python
clf1 = LogisticRegression(random_state=42)
clf2 = RandomForestClassifier(n_estimators=200, random_state=42, max_depth=50)

eclf = VotingClassifier(
     estimators=[('lr', clf1), ('rf', clf2)],
     voting='hard')

for clf, label in zip([clf1, clf2, eclf], ['Logistic Regression', 'Random Forest', 'Ensemble']):
     scores = cross_val_score(clf, x_train, y, scoring='accuracy', cv=5)
     print("Accuracy: %0.2f (+/- %0.2f) [%s]" % (scores.mean(), scores.std(), label))
```

```python
eclf.fit(x_train,y)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(eclf.predict(x_train),y)
```

```python
y_test = eclf.predict(x_test)
y_test = pd.Series(y_test)
y_test = y_test.map({0:'no',1:'yes'})
test_df = pd.read_csv('test.csv')
id_df = test_df['id']
result2 = pd.concat([id_df,y_test],axis=1)
result2
```

```python
result2.columns = ['id','subscribe']
result2.to_csv('result2.csv',index=False)
```
