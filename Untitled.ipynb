```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv('train.csv')
train.head()
```

```python
test = pd.read_csv('test.csv')
test.head()
```

```python
train.drop(columns='id',inplace=True)
train.head()
```

```python
test.drop(columns='id',inplace=True)
test.head()
```

```python
y = train['subscribe'].map({'no':0,'yes':1})
y.head()
```

```python
train.drop(columns='subscribe',inplace=True)
data = pd.concat([train,test],axis=0)
data.head()
```

```python
data.shape
```

```python
num_columns = data.select_dtypes(exclude='O').columns
data[num_columns]
```

```python
from sklearn import preprocessing
normalizer = preprocessing.Normalizer().fit(data[num_columns])
data[num_columns] = normalizer.transform(data[num_columns])
data[num_columns]
```

```python
cat_columns = data.select_dtypes(include='O').columns
data[cat_columns]
```

```python
data = pd.get_dummies(data)
data.head()
```

```python
x_train = data.iloc[train.index]
x_test = data.iloc[test.index]
```

```python
x_train.shape,x_test.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier


clf1 = LogisticRegression(random_state=1)
clf2 = RandomForestClassifier(n_estimators=50, random_state=1)
clf3 = GaussianNB()

eclf = VotingClassifier(
     estimators=[('lr', clf1), ('rf', clf2), ('gnb', clf3)],
     voting='hard')

for clf, label in zip([clf1, clf2, clf3, eclf], ['Logistic Regression', 'Random Forest', 'naive Bayes', 'Ensemble']):
     scores = cross_val_score(clf, x_train, y, scoring='accuracy', cv=5)
     print("Accuracy: %0.2f (+/- %0.2f) [%s]" % (scores.mean(), scores.std(), label))
```

```python
eclf.fit(x_train,y)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(eclf.predict(x_train),y)
```

```python
y_test = eclf.predict(x_test)
```

```python
y_test = pd.Series(y_test)
```

```python
y_test = y_test.map({0:'no',1:'yes'})
```

```python
test_df = pd.read_csv('test.csv')
id_df = test_df['id']
id_df
```

```python
result = pd.concat([id_df,y_test],axis=1)
result
```

```python
result.columns = ['id','subscribe']
```

```python
result
```

```python
result.to_csv('result.csv',index=False)
```
